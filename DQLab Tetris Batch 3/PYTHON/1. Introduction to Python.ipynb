```python
# Adding elements to a list
fruits = ['apple', 'banana', 'orange']
fruits.append('grape')
print(fruits)  # Output: ['apple', 'banana', 'orange', 'grape']
```

```python
# Removing elements from a list
fruits = ['apple', 'banana', 'orange']
fruits.remove('banana')
print(fruits)  # Output: ['apple', 'orange']
```

```python
# indexing a list
fruits = ['apple', 'banana', 'orange', 'grape']
print(fruits[0])  # Output: 'apple'
```

```python
# Slicing a list
fruits = ['apple', 'banana', 'orange', 'grape']
print(fruits[0:2])  # Output: ['apple', 'banana']
# fuits[0], fruits[1]
```

## Dictionary

```python
# Creating a dictionary
person = {"name": "Alice", 
          "age": 30, 
          "occupation": "Engineer"}

person
```

```python
# Accessing values from a dictionary
person = {"name": "Alice", "age": 30, "occupation": "Engineer"}
print(person["name"])  # Output: "Alice"
print(person["age"])  # Output: 30
```

```python
# Adding a new key-value pair to a dictionary
person = {"name": "Alice", "age": 30, "occupation": "Engineer"}
person["location"] = "New York"
print(person)  # Output: {"name": "Alice", "age": 30, "occupation": "Engineer", "location": "New York"}
```

```python
# Modifying a value in a dictionary
person = {"name": "Alice", "age": 30, "occupation": "Engineer"}
person["age"] = 31
print(person)  # Output: {"name": "Alice", "age": 31, "occupation": "Engineer"}
```

```python
# Removing a key-value pair from a dictionary
person = {"name": "Alice", "age": 30, "occupation": "Engineer"}
del person["occupation"]
print(person)  # Output: {"name": "Alice", "age": 30}
```

## Conditional Statements

```python
# if statement
x = 5

if x < 10:
    print("x is less than 10")
```

```python
# if-else statement
x = 15
if x < 10:
    print("x is less than 10")
else:
    print("x is greater than or equal to 10")
```

```python
# if-elif-else statement:
x = 5
if x < 5:
    print("x is less than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is greater than 5")
```

## User-defined Functions

```python
# simple function to add two numbers
def add_numbers(x, y):
    result = x ** y
    return result

# Call the function
sum = add_numbers(2, 3)
print(sum)
```

```python
# Function that returns the absolute value of a number
def abs_val(num):
    if num < 0:
        return -num
    else:
        return num

abs_val(-100)
```
